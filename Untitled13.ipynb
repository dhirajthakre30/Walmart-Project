```python
from prophet import Prophet
```

```python
periods = 12
```

```python
stores = df['Store'].unique()
```

```python
store_forecasts = {}

for store_id in stores:
    store_df = df[df['Store'] == store_id][['Date', 'Weekly_Sales']]
    store_df = store_df.rename(columns={'Date': 'ds', 'Weekly_Sales': 'y'})

    model = Prophet()
    model.fit(store_df)

    future = model.make_future_dataframe(periods=periods, freq='W')
    forecast = model.predict(future)

    store_forecasts[store_id] = forecast[['ds', 'yhat']].tail(periods)
    store_forecasts[store_id]['Store'] = store_id

# Combine forecasts
final_forecast = pd.concat(store_forecasts.values())
final_forecast = final_forecast.rename(columns={'ds': 'Date', 'yhat': 'Predicted_Weekly_Sales'})
```
